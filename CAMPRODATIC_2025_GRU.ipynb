Kode ini membagi data menjadi tiga bagian: data pelatihan (training), validasi, dan pengujian (testing) menggunakan metode Hold-Out Validation. Pertama, panjang data pelatihan ditentukan sebagai 80% dari total data (train_len), kemudian data validasi dihitung sebagai 20% dari data pelatihan (val_len). Sisa dari data pelatihan setelah dikurangi data validasi digunakan sebagai data pelatihan yang sebenarnya. Terakhir, data pengujian (test_len) diambil sebanyak 20% dari total data. Pembagian ini bertujuan untuk memastikan model dapat dilatih, divalidasi, dan diuji secara terpisah untuk menghindari overfitting dan memastikan evaluasi yang lebih objektif.

```python
# Melakukan pembagian data dengan Hold-Out Validation
train_len = math.ceil(len(data_sc) *.8)
val_len = math.ceil(train_len * 0.2)
train_len = train_len - val_len
test_len = math.ceil(len(data_sc) *.2)
```

Kode ini membuat "timesteps" untuk data pelatihan dengan cara membentuk urutan input (X) dan output (Y) yang digunakan dalam model time series. Data pelatihan (train_data) dipilih dari data yang telah dinormalisasi dan dibagi sebelumnya. Kemudian, untuk setiap titik data mulai dari indeks ke-10 hingga akhir, urutan 10 data sebelumnya (train_data[i-10:i, 0]) digunakan sebagai input (x_train), dan nilai data saat ini (train_data[i, 0]) menjadi output yang diprediksi (y_train). Ini memungkinkan model untuk belajar dari 10 data terakhir untuk memprediksi nilai selanjutnya, yang umum diterapkan pada model berbasis urutan seperti jaringan saraf rekuren (RNN).

```python
# Membuat Timesteps pada data Training
train_data = data_sc[0:train_len  , : ]
x_train=[]
y_train = []
for i in range(10, len(train_data)):
    x_train.append(train_data[i-10:i,0])
    y_train.append(train_data[i,0])
```

Kode ini digunakan untuk mengonversi data yang telah dinormalisasi (misalnya, data_sc) ke dalam format yang lebih mudah dipahami, yaitu dalam bentuk tabel dengan "timesteps" sebagai fitur dan nilai target untuk pelatihan model. Dimulai dengan mengambil data pelatihan dari data_sc, kode kemudian membuat urutan data sebelumnya (10 data terakhir) untuk setiap titik data sebagai input (fitur x_train) dan nilai saat ini sebagai output yang diprediksi (target y_train). Hasilnya, x_train dan y_train disusun dalam dua DataFrame terpisah dan kemudian digabungkan menjadi satu tabel yang menyatukan fitur (Timestep_1 hingga Timestep_10) dan target (Target). Tabel ini memudahkan dalam melihat struktur data yang akan digunakan untuk pelatihan model.

```python
import pandas as pd
import numpy as np

# Contoh data yang telah dinormalisasi (gunakan data_sc dari hasil normalisasi sebelumnya)
# Misalkan data_sc sudah ada dari proses sebelumnya
train_data = data_sc[0:train_len, :]

# Membuat timesteps untuk training data
x_train = []
y_train = []

for i in range(10, len(train_data)):
    x_train.append(train_data[i-10:i, 0])
    y_train.append(train_data[i, 0])

# Mengonversi ke dalam DataFrame untuk menampilkan dalam bentuk tabel
x_train_df = pd.DataFrame(x_train, columns=[f'Timestep_{i}' for i in range(1, 11)])
y_train_df = pd.DataFrame(y_train, columns=['Target'])

# Menggabungkan features (timesteps) dan target dalam satu tabel
train_timesteps_df = pd.concat([x_train_df, y_train_df], axis=1)

# Menampilkan tabel timesteps training
print(train_timesteps_df)
```

Kode ini mengonversi data pelatihan (x_train dan y_train) yang awalnya berupa list Python menjadi array NumPy.

```python
# Mengubah tipe data list menjadi NumPy array
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
# Ubah bentuk data menjadi bentuk yang diterima oleh GRU
x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1],1))
```

Kode ini membuat "timesteps" untuk data validasi dengan cara yang sama seperti pada data pelatihan, tetapi kali ini diterapkan pada data validasi (val_data). Dimulai dengan memilih data validasi dari bagian akhir data pelatihan hingga sebagian data berikutnya, kode kemudian membuat urutan 10 data sebelumnya (val_data[i-10:i, 0]) sebagai input (x_val) dan nilai data saat ini (val_data[i, 0]) sebagai output yang diprediksi (y_val). Proses ini memungkinkan model untuk mempelajari pola dari data validasi yang digunakan untuk evaluasi selama pelatihan.

```python
# Membuat Timesteps pada data Validation
val_data = data_sc[train_len - 10:(train_len + val_len) , : ]
x_val = []
y_val = []
for i in range(10,len(val_data)):
    x_val.append(val_data[i-10:i,0])
    y_val.append(val_data[i,0])
```

```python
# Mengubah tipe data list menjadi NumPy array
x_val, y_val = np.array(x_val), np.array(y_val)
```

```python
# Ubah bentuk data menjadi bentuk yang diterima oleh GRU
x_val = np.reshape(x_val, (x_val.shape[0],x_val.shape[1],1))
```

Berikut adalah penjelasan singkat mengenai arsitektur model dengan **GRU**:

Model ini menggunakan **GRU** (Gated Recurrent Unit) untuk prediksi deret waktu, seperti harga saham. Dimulai dengan lapisan GRU pertama (100 unit, `return_sequences=True`) untuk memproses urutan data. Kemudian, ada lapisan **Dropout** (0.2) untuk mengurangi overfitting. GRU kedua (100 unit, `return_sequences=False`) mengolah output dari GRU pertama dan menghasilkan representasi akhir. Lapisan **Dropout** diterapkan lagi sebelum lapisan output **Dense** (1 unit) untuk menghasilkan prediksi tunggal, yang cocok untuk tugas regresi.

**GRU** membantu menangani data urutan dan mengatasi masalah vanishing gradient, lebih efisien dan cepat dibandingkan dengan LSTM.

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import GRU, Dropout, Dense

model = Sequential()
model.add(GRU(units=100, return_sequences=True, input_shape=(x_val.shape[1], x_val.shape[2])))
model.add(Dropout(0.2))
model.add(GRU(units=100))
model.add(Dropout(0.2))
model.add(Dense(units=1))  # Kasus regresi memprediksi satu output
```

Kode ini mengonfigurasi dua fungsi callback untuk melatih model dengan cara yang lebih efisien dan menghindari overfitting. **EarlyStopping** digunakan untuk menghentikan pelatihan jika tidak ada peningkatan pada nilai `val_loss` selama 5 epoch berturut-turut (`patience=5`), dan akan mengembalikan bobot terbaik model (`restore_best_weights=True`). Sementara itu, **ModelCheckpoint** digunakan untuk menyimpan bobot model terbaik yang menghasilkan `val_loss` terendah ke dalam file `weights_bigru_model.weights.h5`. Bobot hanya disimpan jika ada perbaikan pada `val_loss` dengan `save_best_only=True`. Kedua callback ini digabungkan dalam list `callbacks` untuk digunakan selama proses pelatihan model, meningkatkan efisiensi dan menghindari overfitting.

```python
import tensorflow as tf
import os
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Early Stopping
early_stopping = EarlyStopping(
    monitor="val_loss",
    patience=5,
    min_delta=0.00001,
    verbose=1,
    restore_best_weights=True
)

# Model Checkpoint
checkpoint_path = 'model_checkpoint/weights_gru_model.weights.h5'
checkpoint_dir = os.path.dirname(checkpoint_path)
checkpoint = ModelCheckpoint(
    filepath=checkpoint_path,
    monitor='val_loss',
    mode='min',
    save_best_only=True,
    save_weights_only=True,
    verbose=1
)

# Definisikan fungsi callback
callbacks = [checkpoint, early_stopping]
```

```python
# Konfigurasi optimizer dan loss function serta Compile Model
op = tf.keras.optimizers.Nadam(learning_rate=0.001)
model.compile(optimizer=op, loss='mean_squared_error')
```

Kode ini melatih model GRU menggunakan data pelatihan (`x_train` dan `y_train`) selama 25 epoch dengan ukuran batch 16. Selama pelatihan, data validasi (`x_val` dan `y_val`) digunakan untuk mengevaluasi kinerja model setelah setiap epoch. Callback yang telah dikonfigurasi sebelumnya, yaitu **ModelCheckpoint** dan **EarlyStopping**, digunakan untuk menyimpan bobot terbaik model dan menghentikan pelatihan jika tidak ada perbaikan pada nilai `val_loss` dalam 5 epoch berturut-turut. Hasil pelatihan disimpan dalam variabel `history`, yang berisi metrik pelatihan seperti loss dan akurasi pada setiap epoch.

```python
# Melakukan training pada model GRU
history = model.fit(x_train, y_train, batch_size=16, epochs=25, validation_data=(x_val, y_val), callbacks=[checkpoint, early_stopping])
```

```python
# Menampilkan plot nilai loss
plt.figure(figsize=(6,3))
plt.plot(history.history['loss'], linewidth=2)
plt.plot(history.history['val_loss'], linewidth=2)
plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'valid'])
plt.show()
```

Kode ini membuat "timesteps" untuk data pengujian (`test_data`) dengan cara yang sama seperti pada data pelatihan dan validasi. Dimulai dengan memilih data pengujian yang mencakup 10 data terakhir dari total data yang ada, kemudian urutan 10 data sebelumnya (`test_data[i-10:i, 0]`) digunakan sebagai input (`x_test`), dan nilai `Close` dari data pengujian digunakan sebagai output yang diprediksi (`y_test`). Hasilnya adalah dataset `x_test` yang berisi urutan input untuk prediksi, dan `y_test` yang berisi nilai aktual yang digunakan untuk mengukur kinerja model pada data pengujian.

```python
# Membuat Time Step pada data testing
test_data = data_sc[(-test_len - 10): , : ]
#Create the x_test and y_test data sets
x_test = []
y_test =  data[['Close']].iloc[-test_len : , : ].values
for i in range(10,len(test_data)):
    x_test.append(test_data[i-10:i,0])
```

```python
# Mengubah tipe data list menjadi NumPy array
x_test = np.array(x_test)
```

```python
# Ubah bentuk data menjadi bentuk yang diterima oleh GRU
x_test = np.reshape(x_test, (x_test.shape[0],x_test.shape[1],1))
```

```python
# Load bobot model terbaik
model.load_weights("/content/model_checkpoint/weights_gru_model.weights.h5")
```

```python
# Melakukan prediksi
y_pred = model.predict(x_test)
y_pred = np.ceil(sc.inverse_transform(y_pred))
```

Kode ini digunakan untuk mengevaluasi kinerja model menggunakan empat metrik regresi yang umum: MSE (Mean Squared Error), RMSE (Root Mean Squared Error), MAE (Mean Absolute Error), dan MAPE (Mean Absolute Percentage Error). Metrik ini membandingkan hasil prediksi (y_pred) dengan nilai aktual (y_test). MSE mengukur rata-rata kuadrat selisih antara prediksi dan nilai aktual, RMSE adalah akar kuadrat dari MSE, memberikan gambaran yang lebih mudah dimengerti dalam satuan yang sama dengan data asli. MAE mengukur rata-rata selisih absolut antara prediksi dan nilai aktual, sementara MAPE mengukur persentase kesalahan absolut rata-rata. Semua metrik dihitung dan dicetak dengan pembulatan 4 angka desimal untuk memberikan gambaran tentang seberapa baik model dalam memprediksi data.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error

# Menghitung nilai MSE, RMSE, MAE, dan MAPE
mse = round(mean_squared_error(y_test, y_pred), 4)
rmse = round(np.sqrt(np.mean((y_test - y_pred) ** 2)), 4)
mae = round(mean_absolute_error(y_test, y_pred), 4)
mape = round(mean_absolute_percentage_error(y_test, y_pred) * 100, 4)

# Menampilkan hasil evaluasi
print(f'MSE  : {mse}')
print(f'RMSE : {rmse}')
print(f'MAE  : {mae}')
print(f'MAPE : {mape} %')
```
