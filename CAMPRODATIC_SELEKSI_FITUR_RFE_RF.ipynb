```python
import pandas as pd
data = pd.read_excel('Data Kinerja Siswa (1).xlsx')
data.head(10)
```

# **PREPROCESSING DATA**

**Cek Informasi, Nilai Null, dan Duplikasi**

```python
print("\nInformasi Data:")
print(data.info())

# Mengecek nilai null di setiap kolom
null_values = data.isnull().sum()
print("\nJumlah Nilai Null di Setiap Kolom:\n", null_values)

# Menampilkan jumlah dan baris duplikat
duplikat_count = data.duplicated().sum()
print(f"\nJumlah Baris Duplikat: {duplikat_count}")

if duplikat_count > 0:
    print("\nData Duplikat:")
    print(data[data.duplicated(keep=False)])
else:
    print("\nTidak ada data duplikat.")
```

```python
# Hapus kolom 'No' karena tidak relevan
data = data.drop(columns=['No'])
```

# **TRANSFORMASI DATA (Label Encoding)**

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

def convert_to_numeric(data):
    for column in data.columns:
        if data[column].dtype == 'object':
            data[column] = label_encoder.fit_transform(data[column])
    return data

data = convert_to_numeric(data)
print("\nData Setelah Label Encoding:")
print(data.head(10))
```

# **NORMALISASI DATA**

```python
from sklearn.preprocessing import MinMaxScaler

# Pisahkan fitur dan label
X = data.drop(columns=['Prediction'])
y = data['Prediction']

# Inisialisasi scaler
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# Konversi kembali ke DataFrame
X = pd.DataFrame(X_normalized, columns=X.columns)

print("\nData Setelah Normalisasi:")
print(X.head(10))
```

**PEMISAHAN DATA TRAIN-TEST**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print(f"Jumlah Data Latih: {len(X_train)}, Jumlah Data Uji: {len(X_test)}")
```

# **SELEKSI FITUR DENGAN RFE**

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
import pandas as pd

# Inisialisasi Random Forest untuk seleksi fitur dengan RFE
model = RandomForestClassifier(n_estimators=100, random_state=42)
rfe = RFE(model, n_features_to_select=None)  # Biarkan RFE memilih jumlah fitur terbaik
fit = rfe.fit(X_train, y_train)

# Menampilkan hasil seleksi fitur
selected_features = X_train.columns[fit.support_]
print(f"Jumlah Fitur Terpilih: {fit.n_features_}")
print(f"Fitur yang Terpilih: {selected_features.tolist()}")
print(f"Ranking Fitur: {fit.ranking_}")

# Menyimpan fitur terpilih dalam DataFrame untuk proses analisis
selected_features_df = pd.DataFrame({
    'Fitur': X_train.columns,
    'Terpilih': fit.support_,
    'Ranking': fit.ranking_
}).sort_values(by='Ranking')

print("\nRincian Seleksi Fitur:")
print(selected_features_df)

# Transformasi data train dan test menggunakan fitur yang terpilih
X_train_selected = fit.transform(X_train)
X_test_selected = fit.transform(X_test)
```

# **Pelatihan Model Random Forest dengan Seleksi Fitur**

```python
# Inisialisasi model Random Forest setelah seleksi fitur
rf_model = RandomForestClassifier(
    n_estimators=100,   # Jumlah pohon dalam hutan
    max_depth=10,       # Kedalaman maksimum setiap pohon
    min_samples_split=5, # Minimum sampel untuk split
    min_samples_leaf=2,  # Minimum sampel per daun
    random_state=42,
    n_jobs=-1           # Menggunakan semua core CPU
)

# Latih model dengan fitur yang sudah diseleksi
rf_model.fit(X_train_selected, y_train)

# Prediksi data uji
y_pred = rf_model.predict(X_test_selected)

# Evaluasi model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f'\nAkurasi Model: {accuracy * 100:.2f}%')
print(f'Precision: {precision * 100:.2f}%')
print(f'Recall: {recall * 100:.2f}%')
print(f'F1-Score: {f1 * 100:.2f}%')

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

# **Pelatihan Random Forest tanpa Seleksi Fitur**

```python
# Inisialisasi model Random Forest
rf_model = RandomForestClassifier(
    n_estimators=100,   # Jumlah pohon dalam hutan
    max_depth=10,       # Kedalaman maksimum setiap pohon
    min_samples_split=5, # Minimum sampel untuk split
    min_samples_leaf=2,  # Minimum sampel per daun
    random_state=42,
    n_jobs=-1           # Menggunakan semua core CPU untuk mempercepat proses
)

# Latih model menggunakan semua fitur tanpa seleksi
rf_model.fit(X_train, y_train)

# Prediksi data uji
y_pred = rf_model.predict(X_test)

# Evaluasi model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f'\nAkurasi Model: {accuracy * 100:.2f}%')
print(f'Precision: {precision * 100:.2f}%')
print(f'Recall: {recall * 100:.2f}%')
print(f'F1-Score: {f1 * 100:.2f}%')

# Laporan Klasifikasi
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
